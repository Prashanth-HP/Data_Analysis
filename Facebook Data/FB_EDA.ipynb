```python
age_group = df['age_group'].value_counts().reset_index()
age_group.sort_values(by='age_group',inplace=True)
age_group
```

```python
import matplotlib .pyplot as plt
plt.figure(figsize=(12,10))
plt.bar(age_group['age_group'],age_group['count'], color='green')
plt.xlabel('age_group')
plt.ylabel('count')
plt.title("Age vs Count")
plt.show()
```

```python

plt.figure(figsize=(12,10))
plt.plot(age_group['age_group'],age_group['count'], color='green',marker='o')
plt.xlabel('age_group')
plt.ylabel('count')
plt.title("Age vs Count")
plt.show()
```

```python
gender= df['gender'].value_counts().reset_index()
gender
```

```python
gender_count= df['gender'].value_counts()
gender_percentage= df['gender'].value_counts(normalize=True)*100
gender_freq_table=pd.DataFrame({'Count': gender_count,'Percentage':gender_percentage})
gender_freq_table
```

```python
plt.figure(figsize=(10,8))
plt.bar(gender['gender'], gender['count'], color=['b','g'])
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")
plt.show()
```

```python
age_gender_grp = df[['age_group','gender']].value_counts().reset_index().sort_values(by=['age_group', 'gender'], ascending=True).reset_index() # Specify the 'by' argument with the column(s) to sort by
age_gender_grp # Renamed variable to reflect the correct name used later
```

```python
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.barplot(data=age_gender_grp, x='age_group', y='count', hue='gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='age_group', y='likes',hue='gender')
plt.xlabel("Age_grp")
plt.ylabel("Likes")
plt.title("Distribution of likes by AGe Group")
plt.show()
```
