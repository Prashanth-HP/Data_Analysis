```python
import pandas as pd
df  = pd.read_csv("/content/cities air quality water pollution..csv")
df.head()
```

```python
df.info()
```

```python
import plotly.express as px

# Assuming df is your DataFrame and 'country' is the column with country names
data = df['Country'].value_counts().to_frame().reset_index()
data.columns = ['Country', 'Measure']  # Renaming columns for labeling
data = data.sort_values(ascending=False, by='Measure').head(20)

fig = px.bar(data, x='Country', y='Measure', title="Measure by Country")

# Customize layout (if needed)
fig.update_layout(
    xaxis_title="Country",
    yaxis_title="Count",
    title_x=0.5  # Center the title
)

# Show the plot
fig.show()
```

```python
from plotly.express import histogram
for column in df.columns[-3:-1]:
    histogram(data_frame=df, x=column, nbins=100).show()
```

```python
from plotly.express import scatter
scatter(data_frame=df[df['Country'].isin(df['Country'].value_counts()[:10].index.tolist())], x=' WaterPollution', y=' AirQuality', height=600,
        color='Country', title='Data for top 10 countries by datapoint count', hover_name='City')
```

```python
px.bar(data_frame=df[['Country', ' AirQuality']].groupby(by='Country').mean().reset_index().sort_values(ascending=False, by=' AirQuality').head(n=20), x='Country', y=' AirQuality', title='Top 20 countries by mean reported air quality')
```

```python
px.bar(data_frame=df[['Country', ' WaterPollution']].groupby(by='Country').mean().reset_index().sort_values(ascending=False, by=' WaterPollution').head(n=20), x='Country', y=' WaterPollution', title='Top 20 countries by mean reported water pollution')
```
